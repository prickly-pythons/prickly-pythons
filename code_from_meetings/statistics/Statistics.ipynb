# Generating random numbers of a particular distribution
Adapted from http://people.duke.edu/~ccc14/sta-663-2017/15A_Random_Numbers.html

```python
import scipy.stats as stats
```

```python
def expon_pdf(x, lmabd=1):
    """PDF of exponential distribution."""
    return lmabd*np.exp(-lmabd*x)

def expon_cdf(x, lambd=1):
    """CDF of exponetial distribution."""
    return 1 - np.exp(-lambd*x)

def expon_icdf(p, lambd=1):
    """Inverse CDF of exponential distribution - i.e. quantile function."""
    return -np.log(1-p)/lambd
```

```python
x = np.linspace(0,4,100)
```

```python
plt.figure(figsize=(12,4))

plt.subplot(121)
plt.plot(x, expon_cdf(x))
plt.axis([0, 4, 0, 1])
for q in [0.5, 0.8]:
    plt.arrow(0, q, expon_icdf(q)-0.1, 0, head_width=0.05, head_length=0.1, fc='b', ec='b')
    plt.arrow(expon_icdf(q), q, 0, -q+0.1, head_width=0.1, head_length=0.05, fc='b', ec='b')
plt.ylabel('1: Generate a (0,1) uniform PRNG')
plt.xlabel('2: Find the inverse CDF')
plt.title('Inverse transform method');

plt.subplot(122)
u = np.random.random(10000)
v = expon_icdf(u)
plt.hist(v, histtype='step', bins=100, normed=True, linewidth=2)
plt.plot(x, expon_pdf(x), linewidth=2)
plt.axis([0,4,0,1])
plt.title('Histogram of exponential PRNGs')

plt.show()
```

# Wisdom of the Crowd

## Q1: What is the current population of Denmark?

```python
num_bin = 5

guess = np.array([10e6, 20e6, 12.5e6, 5e6, 30e6])
expert = 5.5e6
guess_mean = np.mean(guess)

print("mean   = %.2e" % guess_mean)
print("median = %.2e" % np.median(guess))
print("min    = %.2e" % np.min(guess))
print("max    = %.2e" % np.max(guess))
print("std    = %.2e" % np.std(guess))
```

```python
true_val = 5.7e6

fig = plt.figure(figsize=(12,5))
ax = fig.add_subplot(111)
ax.hist(guess, bins=num_bin)
ax.axvline(x=true_val, color='r', linestyle='--', linewidth=3)
ax.axvline(x=guess_mean, color='m', linestyle='-', linewidth=3)
ax.axvline(x=expert, color='g', linestyle='-', linewidth=3)
plt.show()
```
